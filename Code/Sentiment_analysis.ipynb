```python

# Sentiment Analysis using IMDb Dataset
# Run in Google Colab

import pandas as pd
import numpy as np

from sklearn.datasets import fetch_openml
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Download IMDb Dataset
print("Loading Dataset...")

from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.datasets.imdb import get_word_index

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=10000)

word_index = get_word_index()

reverse_word_index = {
    value + 3: key for (key, value) in word_index.items()
}
reverse_word_index[0] = "<PAD>"
reverse_word_index[1] = "<START>"
reverse_word_index[2] = "<UNK>"
reverse_word_index[3] = "<UNUSED>"

def decode_review(text):
    return " ".join([reverse_word_index.get(i, "?") for i in text])

train_reviews = [decode_review(review) for review in X_train[:10000]]
test_reviews = [decode_review(review) for review in X_test[:2500]]

print("Dataset Loaded Successfully")

# TF-IDF Vectorization
vectorizer = TfidfVectorizer(
    max_features=10000,
    stop_words='english'
)

X_train_tfidf = vectorizer.fit_transform(train_reviews)
X_test_tfidf = vectorizer.transform(test_reviews)

# Logistic Regression Model
model = LogisticRegression(max_iter=1000)

print("Training Model...")
model.fit(X_train_tfidf, y_train[:10000])

# Prediction
predictions = model.predict(X_test_tfidf)

accuracy = accuracy_score(y_test[:2500], predictions)

print("\nAccuracy =", round(accuracy*100,2), "%")

print("\nClassification Report:\n")
print(classification_report(y_test[:2500], predictions))

# User Testing
try:
    while True:

        review = input("\nEnter Review (type exit to stop): ")

        if review.lower() == "exit":
            break

        review_vector = vectorizer.transform([review])

        result = model.predict(review_vector)[0]

        if result == 1:
            print("Sentiment: Positive 😊")
        else:
            print("Sentiment: Negative 😞")
except KeyboardInterrupt:
    print("\nOperation interrupted by user. Exiting.")
```
